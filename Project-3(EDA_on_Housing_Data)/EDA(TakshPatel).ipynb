# Perform EDA using SEABORN,MATPLOTLIB on housing dataset.

```python
!pip install seaborn
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Housing.csv')
df
```

```python
sns.set(style="whitegrid", palette="muted")
```

```python
print("Dataset Shape:", df.shape)
print("\nDataset Info:\n")
df.info()
print("\nSummary Statistics:\n")
df.describe()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['price'], kde=True, bins=30, color="blue")
plt.title("Distribution of House Prices")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="furnishingstatus", data=df, palette="Set2")
plt.title("Furnishing Status Distribution")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(x="bedrooms", y="price", data=df, palette="pastel")
plt.title("Price Distribution by Bedrooms")
plt.show()
```

```python
sns.pairplot(df[["price","area","bedrooms","bathrooms","stories","parking"]], diag_kind="kde")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
